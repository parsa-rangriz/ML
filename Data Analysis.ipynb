# Regression

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors
n_neighbors = 15

X_data = [data.y, data.x91, data.x92]
X = np.zeros((len(X_data[0]), len(X_data)))
for i in range(len(X_data)):
    for j in range(len(X_data[i])):
        X[j][i] = X_data[i][j]
```

## Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
regr_1 = DecisionTreeRegressor(max_depth=5)
regr_2 = DecisionTreeRegressor(max_depth=5)
X0 = np.array([X[:, 0]]).T
X1 = np.array([X[:, 1]]).T
X2 = np.array([X[:, 2]]).T
regr_1.fit(X0, X1)
regr_2.fit(X0, X2)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
plt.figure()
plt.scatter(X0, X1, s=20, label="x91")
plt.scatter(X0, X2, s=20, label="x92")
plt.plot(X_test, y_1, label="max_depth=5", c = 'yellow')
plt.plot(X_test, y_2, label="max_depth=5", c = 'blue')
plt.legend()
plt.show()
```

## Nearest Neighbors

```python
from sklearn import neighbors
T = np.linspace(0, 5, 500)[:, np.newaxis]
n_neighbors = 5
knn = neighbors.KNeighborsRegressor(n_neighbors, weights = 'uniform')
y_ = knn.fit(X0, X1).predict(T)
y__ = knn.fit(X0, X2).predict(T)
plt.scatter(X0, X1, color='darkorange', label='data')
plt.scatter(X0, X2, color='blue', label='data')
plt.plot(T, y_, color='navy', label='prediction')
plt.plot(T, y__, color='orange', label='prediction')
plt.legend()                                                   
plt.show()
```
